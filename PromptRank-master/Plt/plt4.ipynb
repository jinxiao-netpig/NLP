```python
from utils.logger import LogReader
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lr = LogReader()
log_dir1  = '../Logs/all'
log_dir2 = '../Logs/peter'

data1 = lr.read_logs(log_dir1)
data2 = lr.read_logs(log_dir2)  # 读取日志
df1   = pd.DataFrame(data1)
df2   = pd.DataFrame(data2)

df = pd.concat([df1, df2], axis=0)
df = df.loc[df['dataset'] == 'TripAdvisor'][:-1]  # 去除最后一次异常数据
df
```

```python
df = df.groupby('model').mean(numeric_only=True).reset_index()
df
```

```python
# 选择数值列
numeric_cols = df.select_dtypes(include='number').columns
# 选择基准行，假设基准行为索引为 2 的行
reference_row = df.loc[0, numeric_cols]
# 计算所有行与基准行的数值差
difference = df[numeric_cols] - reference_row
# 计算差值的百分比
percentage_difference = (difference / reference_row) * 100

res = pd.concat([df['model'], percentage_difference], axis=1)
res
```
