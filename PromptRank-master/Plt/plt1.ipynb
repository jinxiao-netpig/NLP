```python
from utils.logger import LogReader
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lr = LogReader()
log_dir = '../Logs/script1'

data = lr.read_logs(log_dir)  # 读取日志
```

```python
df = pd.DataFrame(data)
# [0.01, 0.05, 0.1, 0.2, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0]
# df = df[(df['rating_reg'] >= 0) & (df['rating_reg'] <= 5)]

df
```

```python
paper = {
    'RMSE': 0.81,
    'MAE': 0.63,
    'FMR': 0.07,
    'FCR': 0.13,
    'DIV': 2.95,
    'USR': 0.08,
    'BLEU-1': 15.96,
    'BLEU-4': 1.11,
    'R1-P': 19.07,
    'R1-R': 16.09,
    'R1-F': 16.48,
    'R2-P': 2.33,
    'R2-R': 2.17,
    'R2-F': 2.09,
}

x = 'rating_reg'
y_list = [
    'RMSE', 'MAE', 'DIV',
    'FMR', 'FCR', 'USR',
    'BLEU-1', 'BLEU-4',
    'R1-P', 'R1-R', 'R1-F', 'R2-P', 'R2-R', 'R2-F',
]
```

```python
for y in y_list:
    # 使用 pivot 方法重塑数据，以标题作为列，时间戳作为索引
    df_pivot = df.pivot(index=x, columns='model', values=y)

    # 绘制折线图
    df_pivot.plot(kind='line', marker='o', figsize=(10, 6))
    plt.axhline(y=paper[y], color='r', linestyle='--', label='PETER-Paper')

    # 设置图形属性
    plt.title(y)
    plt.xlabel(x)
    plt.ylabel(y)
    plt.grid(True)
    plt.legend(title='Model')
    plt.xticks(rotation=45)
    plt.tight_layout()

    # 显示图形
    plt.show()
```

实验结果分析

1. PETER-SCoR 和 PETER-MF 是一开始跑的，batch_size 为 128
   
   PETER 是后来加的实验，batch_size 为1024，可能对实验结果有所影响

2. rating_reg 的取值为 [0.01, 0.05, 0.1, 0.2, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0]
   
   在通常情况下，rating_reg 过高都会导致效果下降
   
   通过日志分析，rating_reg 为 0.05 时，评分预测任务已经出现了 **过拟合**

3. 有些实验的结果可能在误差范围内
